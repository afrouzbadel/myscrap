**BTC_USD**


```python

import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://finance.yahoo.com/quote/BTC-USD/history?p=BTC-USD'
page = requests.get(url)
bs = BeautifulSoup(page.content)
table = bs.find('table',{'class':'W(100%) M(0)'})
table_rows = table.find_all('tr')

res = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)


df = pd.DataFrame(res,columns = ['Date','Open','High','Low','Close','Adj Close','Volume'])
# close =df['Close']
# print(close)

print(df)

if df.iloc[-1,1] is None:
    df.drop(df.index[-1],inplace=True)
df['Date']=pd.to_datetime(df['Date'],format='%b %d, %Y')
for name in df.columns:
    if name == 'Date' :
        continue
    df[f'{name}']=df[f'{name}'].apply(lambda x:float(x.replace(',','')))
df.reset_index(inplace=True,drop=True)

plt.figure(figsize=(20,10))
plt.title('Chainlink [Close]Price Over Time')
plt.xlabel('Time')
plt.ylabel('Close Price')

sns.lineplot(data=df,x='Date',y='Close',lw=5,color='blue')
```
